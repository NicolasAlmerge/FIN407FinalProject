## Predictor

This parses a given predictor, given its short name.

```python
values, ports = DailyPortfolios.read_predictor("BidAskSpread", vw=False)
print(ports)
pd.DataFrame(values)
```

```python
values, ports = DailyPortfolios.read_predictor("BidAskSpread", vw=True)
print(ports)
pd.DataFrame(values)
```
